```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as scipy

T=10
M=2000

X = np.random.rand(T,M)<0.3
```

```python
avgs = np.mean(X,0)
```

```python
counts, bins = np.histogram(avgs)
plt.hist(bins[:-1], bins, weights=counts)
```
